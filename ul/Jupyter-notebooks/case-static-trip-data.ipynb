# Case 1: Exploring the trip data

It seems like trip_id's change over the span of a month. In order to not miss any new, removed or changed trips, we need to find a way to store all unique trips and their underlying stops.

* We store trip data from the 1st of January and append all stops for each trip in a python dict
* Then we loop through all the trip data from 2nd to 31st of January to find new trip_id's

```python
import csv
import os

stopsOnTrip = {}

directory = '../data/static/'

with open('../data/static/1/stop_times.txt') as file:
        reader1 = csv.DictReader(file)

        # Loops through stop_times.csv and creates dictionary with stops per unique trip
        for row in reader1:
            stopsOnTrip.setdefault(row['trip_id'],[]).append(row)

newTrips = {}            

for day in os.listdir(directory):
    print("Reading directory",day)
    
    with open('../data/static/' + day + '/stop_times.txt') as file:
        reader1 = csv.DictReader(file)

        # Loops through stop_times.csv and creates dictionary with stops per unique trip
        for row in reader1:
            if row['trip_id'] not in stopsOnTrip:
                newTrips.setdefault(day,[row['trip_id']]).append((row))
                print(day,row['trip_id'])
```
