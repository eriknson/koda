## MAXFLOW interpretation – non_reachable distribution

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

d = pd.read_csv('mincut_values_from_9022003700600.csv')
non_reachable = d['non_reachable']
sink_nam = ['sink_name']
plt.hist([non_reachable], color=['red'],bins = 25, range=(200,3000))
plt.title('non_reachable distribution, range=[200,3000]')
plt.xlabel('non_reachable nodes (stations)')
plt.ylabel('count')
plt.show()
```

```python
non_reachable.describe()
```

## MAXFLOW interpretation – cut_value distribution

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

d = pd.read_csv('mincut_values_from_9022003700600.csv')
cut_value = d['cut_value']
sink_nam = ['sink_name']
plt.hist([cut_value], color=['blue'],range=(0,600),bins = 50)
plt.title('cut_value distribution, range=[0,600]')
plt.xlabel('cut_value for s-t paths')
plt.ylabel('count')
plt.show()
```
