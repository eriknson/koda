```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

d = pd.read_csv('ul-edge-flows.csv')
flow = d['flow']
legend = ['flow']
plt.hist([flow], color=['orange'])
plt.xlabel("Runs/Delivery")
plt.ylabel("Frequency")
plt.legend(legend)
plt.xticks(range(0, 10))
plt.yticks(range(0, 30))
plt.title('Flow distribution')
plt.show()
```
