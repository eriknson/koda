```python
# manual adding of all edges 
G.add_weighted_edges_from(route1) #line 1
G.add_weighted_edges_from(route2) #line 2
G.add_weighted_edges_from(route3) #line 3

D.add_weighted_edges_from(route1back)
D.add_weighted_edges_from(route2back)
D.add_weighted_edges_from(route3back)
```

```python
pos = nx.spring_layout(G)  # positions for all nodes
```

```python

# nodes
nx.draw_networkx_nodes(G, pos, node_size=600)

# edges outgoing routes
nx.draw_networkx_edges(G, pos, edgelist=route1,
                       width=2, edge_color='r')
nx.draw_networkx_edges(G, pos, edgelist=route2,
                       width=2, edge_color='b')
nx.draw_networkx_edges(G, pos, edgelist=route3,
                       width=2, edge_color='g')
# edges ingoing routes
nx.draw_networkx_edges(G, pos, edgelist=route1back,
                       width=2, edge_color='r')
nx.draw_networkx_edges(G, pos, edgelist=route2back,
                       width=2, edge_color='b')
nx.draw_networkx_edges(G, pos, edgelist=route3back,
                       width=2, edge_color='g')

# edges ingoing routes
nx.draw_networkx_edges(D, pos, edgelist=route1back,
                       width=2, edge_color='r')
nx.draw_networkx_edges(D, pos, edgelist=route2back,
                       width=2, edge_color='b')
nx.draw_networkx_edges(D, pos, edgelist=route3back,
                       width=2, edge_color='g')
# node labels
nx.draw_networkx_labels(G, pos, font_size=14, labels=node_names )

# adding edge labels as weights
nx.draw_networkx_edge_labels(G,pos,edge_labels=nx.get_edge_attributes(G,'weight'), label_pos=0.8)

nx.draw_networkx_edge_labels(D,pos,edge_labels=nx.get_edge_attributes(D,'weight'), label_pos=0.8)


plt.show()

plt.rcParams["figure.figsize"] = (15,10)
```

Each edge in the graph G above is directed in both directions and the weight is represented by the number close to the source node. 

The min-cut max-flow theorem states that for a given flow-network, the maximum ammount of flow passing from the source to the sink is equal to the weight of the minimum cut.

```python
#applying the min-cut algorithm to the graph G with defined source (v1) and sink (v10)
nx.minimum_cut(G, v1, v10, capacity = 'weight')
```

The output shows the following three parameters: (value of the cut, {partition 1}, {partition 2})
* Value of the cut = Sum of the weights of the edges where the cut is made (=max flow)
* Partition 1 = Set of nodes that are isolated by the cut 
* Partition 2 = Set of nodes that are isolated by the cut 

The value of the cut is 6 and the partition is made by having v9=(9,5) and v10=(10,5). If we walk from v1 to v10, we see that there are three possible ingoing paths to the partition {(9, 5), (10, 5)} which is:
* The blue line leaving v7 with outgoing weight = 1
* The green line leaving v6 with outgoing weight = 3
* The red line leaving v8 with outgoing weight = 2
Total flow of 6 = value of the cut = max flow passing from v1(soruce) to v10(sink)
