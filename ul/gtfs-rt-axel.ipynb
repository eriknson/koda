```python
from google.transit import gtfs_realtime_pb2
import requests

tripIds = []
feed = gtfs_realtime_pb2.FeedMessage()
response = requests.get('https://opendata.samtrafiken.se/gtfs-rt/ul/VehiclePositions.pb?key=08782a0c3e714b97a4b63f77700dbeb3')
feed.ParseFromString(response.content)
print (feed)
for entity in feed.entity:
    
    tripIds.append(entity.vehicle.trip.trip_id)

print(entity)
    
```
